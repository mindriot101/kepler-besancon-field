## A study of the Besançon model

The Besançon galaxy model was queried to study the effect of the selection cuts made. The constraints used for the Galaxy model were:

* l = 76.32
* b = 13.5
* solid angle: 105 square degrees
* all spectral classes and types
* $5 \leq V \leq 20$

### Kepler selection cuts

* $4.0 \leq \log{g} \leq 4.9$
* $4100 \leq T_{eff} \leq 6100$ K
* $V < 15.1$

*note 15.1 roughly corresponds to $K_p = 15$*

### WASP selection cuts

* $9.3 \leq V \leq 12.8$
* $0.16 \leq \mathrm{J-H} \leq 0.49$
* Dwarfs only (`cl = 5`)

*The last point here is a bit of a cheat. We apply a dwarf probability cut from the proper motion information. The point is to achieve the same cut - only selecting dwarfs - but with the available information.*

## Code

```python
# Imports
%pylab inline
from IPython.display import Image
from pandas.io import sql as psql
import pandas as pd
import psycopg2
```

```python
# Extract the data (warning, uses lots of memory)
with psycopg2.connect(database='besancon') as connection:
    wasp = psql.read_sql('''select * from wasp_sample''', connection)
    kepler = psql.read_sql('''select * from kepler_sample''', connection)
    
    # Only pick a subsample of the full sample as it's quite a lot to plot
    full_hr = psql.read_sql('''select * from kepler_field 
            where cl <= 5
            and id % 10 = 0''', connection)

for dataset in [wasp, kepler, full_hr]:
    dataset['teff'] = 10 ** dataset.ltef
```

```python
print "WASP Objects: {}".format(len(wasp))
print "Kepler Objects: {}".format(len(kepler))
print "Total objects: {}".format(len(full_hr))
```

```python
print "WASP acceptance fraction: {:.3f}%".format(100 * float(len(wasp)) / float(len(full_hr)))
print "Kepler acceptance fraction: {:.3f}%".format(100 * float(len(kepler)) / float(len(full_hr)))
```

```python
# Some helper functions
def plot_histogram(dataset, colour, name):
    columns = ['dist', 'typ', 'metal', 'age']
    dataset.hist(column=columns, fc=colour, bins=15)

def plot_hr_diagram(dataset, colour, name):
    figure(figsize=(8, 4))
    xdata = dataset.teff
    ydata = dataset.mv

    scatter(full_hr.teff, full_hr.mv, marker='.', alpha=0.3,
            c='k', lw=0, zorder=-10)
    scatter(xdata, ydata, marker='.', c=colour, lw=0)
    ax = plt.gca()
    ax.set_xscale('log')
    ax.set_xlim(2E4, 2E3)
    ax.invert_yaxis()
    ax.set_xlabel(r'Teff / K')
    ax.set_ylabel(r'Absolute magnitude')
    ax.set_title(name.replace("_", " ").capitalize())
    tight_layout()
    savefig('img/{}_hr.png'.format(name.capitalize()))
```

WASP tends to observe younger stars. If the age is quoted in Gyr then typically WASP looks at younger stars than the Sun, whereas Kepler looks at older stars. Both samples span the solar age but peak at different points.

### Distance

Both distance distributions peak at around 0.5 parsecs, but the Kepler distribution extends much further (to 2 rather than 1). Kepler therefore observes a high number of more distant stars.

### Metallicity

We plot the metallicity distribution comparisons separately as they require a higher bin resolution than the others.

```python
print "Kepler summary:"
print kepler.metal.describe()
print '---\nWASP summary:'
print wasp.metal.describe()
```

```python
kepler_metal, edges = histogram(kepler.metal, bins=50, range=(-1.5, 1.0))
wasp_metal, _ = histogram(wasp.metal, bins=edges, range=(-1.5, 1.0))

kepler_metal = append(kepler_metal, 0)
wasp_metal = append(wasp_metal, 0)
```

```python
l1 = plot(edges, kepler_metal, 'r-', drawstyle='steps-post', label='Kepler')[0]
ylabel(r'Kepler')
twinx()
l2 = plot(edges, wasp_metal, 'b-', drawstyle='steps-post', label='WASP')[0]
ylabel(r'WASP')
legend([l1, l2], ['Kepler', 'WASP'], loc='upper left')
axvline(wasp.metal.median(), color='b', ls=':')
axvline(kepler.metal.median(), color='r', ls=":")
```

```python
total, edges = histogram(full_hr.metal, bins=30)
kepler_metal, _ = histogram(kepler.metal, bins=edges)
wasp_metal, _ = histogram(wasp.metal, bins=edges)
kepler_metal_fraction = kepler_metal.astype(float) / total
wasp_metal_fraction = wasp_metal.astype(float) / total
plot(edges[:-1], kepler_metal_fraction, 'r-', drawstyle='steps-post')
plot(edges[:-1], wasp_metal_fraction, 'b-', drawstyle='steps-post')
```

```python
print "WASP median metallicity: {}".format(wasp.metal.median())
print "Kepler median metallicity: {}".format(kepler.metal.median())
```

## Applying the Fischer relation

These numbers are just for a single field (large for WASP) and are the underlying rates i.e. they may not be transiting planets.

```python
fischer = lambda metal: 0.03 * 10 ** (2. * metal)
```

```python
n_kepler_planets = fischer(kepler.metal).sum()
n_wasp_planets = fischer(wasp.metal).sum()
print "{} Kepler planets".format(n_kepler_planets)
print "{} WASP planets".format(n_wasp_planets)
```
